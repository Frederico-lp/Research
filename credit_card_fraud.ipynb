```python
import numpy as np

def get_nearest(src_point, candidates):
    min_dist = 1000
    min_index = 0
    for i in range(len(candidates)):
        #distance = dist(src_point, candidates.iloc[[i]])
        dist = np.linalg.norm(src_point - candidates.iloc[[i]])
        print(dist)
        if dist < min_dist:
            min_dist = dist
            min_index = i

    return min_index, min_dist
```
